```python
link = 'https://www.foody.vn/ho-chi-minh/food/an-vat-via-he'

driver.get(link)
```

```python
list_link_rest = []
```

```python
# Đăng nhập trước khi scroll xuống

try:
    while(1):
        if(len(driver.find_elements(by=By.XPATH, value='//div[@class="btn-load-more full-width"]')) != 0):
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            sleep(1)
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            sleep(1)
            driver.find_element(by=By.XPATH, value='//div[@class="btn-load-more full-width"]').click()
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            sleep(1)
except:
    pass

print("---stop---")
```

```python
# Lay tat ca link cua cac cua hang (co the chua chuoi cua hang)
xpath_link_rests = driver.find_elements(by=By.XPATH, value='//div[@class="ri-avatar result-image"]/a')
len(xpath_link_rests)
```

```python
link_total = []
# Chạy vòng for lấy link
for xpath_link_rest in xpath_link_rests:
    link_rest = xpath_link_rest.get_attribute('href')
    link_total.append(link_rest)
```

```python
len(link_total)
```

```python
# Chạy vòng for để lấy các link -> check link luôn trong vòng for
for link in link_total:
    try:
        driver.get(link)
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        sleep(1)

        # check xem có title hay không
        # có thì sẽ là link của cửa hàng
        if(len(driver.find_elements(by=By.XPATH, value='//div[@class="main-info-title"]')) != 0):
            list_link_rest.append(link_rest)
        # không có thì là link của chi nhánh mẹ -> lấy link chi nhánh con
        else:
            xpath_link_rest_seconds = driver.find_elements(by=By.XPATH, value='//div[@class="ldc-item-img"]/a')
            for xpath_link_rest_second in xpath_link_rest_seconds:
                list_link_rest.append(xpath_link_rest_second.get_attribute('href'))

        print('Da lay duoc - ' + str(len(list_link_rest)))
    except:
        continue
```
